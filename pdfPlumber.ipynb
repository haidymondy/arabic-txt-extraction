```python

def reverse_line(line):
    reversed_line = ""
    i = 0
    while i < len(line):
        if line[i].isdigit():
            start = i
            while i < len(line) and line[i].isdigit():
                i += 1
            reversed_line += line[start:i][::-1]
        elif line[i] == '(':
            reversed_line += ')'
            i += 1
        elif line[i] == ')':
            reversed_line += '('
            i += 1
        else:
            reversed_line += line[i]
            i += 1
    return reversed_line[::-1]

with pdfplumber.open("/content/Test (1).pdf") as pdf:
    text = ""
    for page in pdf.pages:
        text += page.extract_text()

reversed_lines = [reverse_line(line) for line in text.split('\n')]
reversed_text1 = '\n'.join(reversed_lines)

print(reversed_text1)
```

```python

with open("reversed_output1.txt", "w") as file:
    file.write(reversed_text1)
```

```python


with open("/content/reversed_output1.txt", "r") as extracted_file:
    extracted_text = extracted_file.read()

with open("/content/Test (1).txt", "r") as original_file:
    original_text = original_file.read()


def f_score(extracted_text, original_text):
    common_chars = sum(1 for char in extracted_text if char in original_text)
    precision = common_chars / max(len(extracted_text), 1)
    recall = common_chars / max(len(original_text), 1)
    if precision + recall == 0:
        return 0
    return 2 * (precision * recall) / (precision + recall)

fs = f_score(extracted_text, original_text)

print("F-score:", fs)
```

```python

def convert_to_arabic_numbers(text):
    arabic_numbers = {'0': '٠', '1': '١', '2': '٢', '3': '٣', '4': '٤',
                      '5': '٥', '6': '٦', '7': '٧', '8': '٨', '9': '٩'}
    converted_text = ""
    for char in text:
        if char.isdigit():
            converted_text += arabic_numbers[char]
        else:
            converted_text += char
    return converted_text

def reverse_line(line):
    reversed_line = ""
    i = 0
    while i < len(line):
        if line[i].isdigit() or line[i] == '.':
            start = i
            while i < len(line) and (line[i].isdigit() or line[i] == '.'):
                i += 1
            reversed_line += line[start:i][::-1]
        elif line[i] == '¶':
            reversed_line += '¶'
        else:
            reversed_line += line[i]
        i += 1
    return reversed_line[::-1]

with pdfplumber.open("/content/Test (2).pdf") as pdf:
    text = "".join(page.extract_text() for page in pdf.pages)

converted_text = convert_to_arabic_numbers(text)
reversed_lines = [reverse_line(line) for line in converted_text.split('\n')]
reversed_text = '\n'.join(reversed_lines)

print(reversed_text)
```
